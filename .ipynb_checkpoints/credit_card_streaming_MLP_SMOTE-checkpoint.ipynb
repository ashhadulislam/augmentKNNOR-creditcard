```python
def check_cm_others(y_actual,y_predict):
    from sklearn.metrics import confusion_matrix

    cm1 = confusion_matrix(y_actual,y_predict)
    print('Confusion Matrix : \n', cm1)

    total1=sum(sum(cm1))
    #####from confusion matrix calculate accuracy
    accuracy1=(cm1[0,0]+cm1[1,1])/total1
    print ('Accuracy : ', accuracy1)

    tn=cm1[0,0]
    fp=cm1[0,1]
    fn=cm1[1,0]
    tp=cm1[1,1]

    sensitivity1 = tp/(tp+fn)
#     print('Sensitivity : ', sensitivity1 )

    specificity1 = tn/(tn+fp)
#     print('Specificity : ', specificity1)
    recall=sensitivity1
    precision=tp/(tp+fp)

#     print("Precision = ",precision)
#     print("Recall = ",recall)

    f1_score=2*(precision*recall)/(precision+recall)
    
    g_mean=math.sqrt(sensitivity1*specificity1)

#     print("F1 score = ",f1_score)

    return accuracy1,sensitivity1,specificity1,precision,recall,f1_score,g_mean

```

```python
df=pd.read_csv("data/creditcard.csv")
```

```python
scaler = StandardScaler()
df['NormalizedAmount'] = scaler.fit_transform(df['Amount'].values.reshape(-1, 1))
df = df.drop(['Amount'], axis = 1)
```

```python
df1 = df.pop('Class') 
df['Class']=df1 

```

```python
df.head()
```

```python
X=df.iloc[:,:-1]
y=df.Class
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.4, random_state=10)
```

### Model to check on non augmented data

```python
X_train_no_aug = X_train.drop(['Time'], axis = 1)
X_test_no_aug = X_test.drop(['Time'], axis = 1)
```

```python
X_train_no_aug.shape
```

```python
model = Sequential([
Dense(input_dim = X_train_no_aug.shape[1], units = 16, activation = 'relu'),
Dense(units = 24, activation = 'relu'),
Dropout(0.5),
Dense(units = 20, activation = 'relu'),
Dense(units = 24, activation = 'relu'),
Dense(units =1, activation = 'sigmoid'),])
```

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
model.fit(X_train_no_aug, y_train, batch_size = 15, epochs = 80)
```

```python
print(model.evaluate(X_test_no_aug,y_test))
```

```python
y_predict=model.predict_classes(X_test_no_aug)
acc,sens,spec,prec,recall,f1_score,g_mean=check_cm_others(y_test,y_predict)


#                 print("Sensitivity = ",sens," Specificity = ",spec," precision=",prec," recall = ",recall," f1_score = ",f1_score," g_mean = ",g_mean )
print(" precision=",prec," recall = ",recall," f1_score = ",f1_score," g_mean = ",g_mean)
```

```python
start=0
total_augmented=0
counts=0
for i in range(0,rows+1,10000):
    start=i
    stop=min((start+10000),rows)
#     print(start,stop)
    X_y_train_batch=X_y_train.iloc[start:stop,:]
    # augmentation
    minority_count=sum(X_y_train_batch.Class==1)
    majority_count=sum(X_y_train_batch.Class==0)    
    imb_ratio=minority_count/majority_count
    print("IR = ",imb_ratio)
    print(X_y_train_batch.Class.value_counts())

    
    if sum(X_y_train_batch.Class==1)>5:
        #remove the time column here
        X_y_train_batch = X_y_train_batch.drop(['Time'], axis = 1)


        
        N=minority_count*2
        print("Augmenting")
        daug = data_augment.DataAugment()
        [Data_a,Ext_d,Ext_not]=daug.augment(data=X_y_train_batch.values,k=3,class_ind=1,N=N,
                                                    randmx=0.0001,dist_percent=0.9)
        print("DOne augmenting")
        total_augmented+=N
        X_train_knnor=Data_a[:,:-1]
        y_train_knnor=Data_a[:,-1]
        model.fit(X_train_no_aug, y_train, batch_size = 15, epochs = 5)
        counts+=1
        print("After KNNOR Majority = ",sum(y_train_knnor==0)," Minority = ",sum(y_train_knnor==1))
        
        print("*****")
```
